```
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

**CREATING SUBPLOTS**

```
fig, ax = plt.subplots(2,3,figsize=(8,5))
fig.tight_layout()
```

**CREATING SUBPLOTS2GRID**

```
figure = plt.figure(figsize=(8,6))

axs = [None for i in range(6)]

axs[0] = plt.subplot2grid((3,4),(0,0), colspan=4)
axs[1] = plt.subplot2grid((3,4),(1,0), colspan=1)
axs[2] = plt.subplot2grid((3,4),(1,1), colspan=1)
axs[3] = plt.subplot2grid((3,4),(1,2), colspan=1)
axs[4] = plt.subplot2grid((3,4),(1,3), colspan=1, rowspan=2)
axs[5] = plt.subplot2grid((3,4),(2,0), colspan=3)

for i in range(6):
  axs[i].set_title("Plot{}".format(i))
  axs[i].set_xticks([])
  axs[i].set_yticks([])

plt.tight_layout()
```

**Practice**

```
x1 = random.randint(10,100,10)
x2 = random.randint(20,200,10)

x3 = random.randint(10,100,10)
x4 = random.randint(20,200,10)

x5 = random.randint(10,100,10)
x6 = random.randint(20,200,10)

x7 = random.randint(10,100,10)
x8 = random.randint(20,200,10)
```

```
figure = plt.figure(figsize=(9,7))

ax = [None for i in range(4)]

ax[0] = plt.subplot2grid((3,4),(0,0),colspan=4)
ax[0].plot(x1, label="x1")
ax[0].plot(x2, label="x2")
ax[0].legend(title="Values")
ax[0].grid()

ax[1] = plt.subplot2grid((3,4),(1,0),colspan=2)
ax[1].plot(x3, label="x3")
ax[1].plot(x4, label="x4")
ax[1].legend(title="Values")
ax[1].grid()

ax[2] = plt.subplot2grid((3,4),(1,2),colspan=2)
ax[2].plot(x5, label="x5")
ax[2].plot(x6, label="x6")
ax[2].legend(title="Values")
ax[2].grid()

ax[3] = plt.subplot2grid((3,4),(2,0),colspan=4)
ax[3].plot(x7, label="x7")
ax[3].plot(x8, label="x8")
ax[3].legend(title="Values")
ax[3].grid()


plt.tight_layout()
```
