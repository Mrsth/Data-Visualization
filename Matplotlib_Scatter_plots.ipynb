```
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```
x = random.randint(1,100, 50)
y = random.randint(1,200, 50)
```

```
plt.figure(figsize=(10,10))

colors = random.randint(1,200,50)

plt.scatter(x,y, s = 100, c=colors,cmap="Greens", alpha=0.85)
plt.title("Scatter plot", fontdict={'fontsize':30})
plt.xlabel("X-axis", fontdict={'fontsize':20})
plt.ylabel("Y-axis", fontdict={'fontsize':20})
cbar = plt.colorbar()
cbar.set_label("Color meter", fontdict={'fontsize':20})
```
