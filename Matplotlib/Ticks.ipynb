```
import matplotlib.pyplot as plt
import numpy as np
from numpy import random 
```

```
array1 = random.randint(10,100,20)
array2 = random.randint(10,200,20)
array3 = random.randint(200,300,20)
```

```
plt.figure(figsize=(20,5))
plt.title("Practice graph 1", fontdict={"fontsize":30})
plt.xlabel("X-axis", fontdict={"fontsize":20})
plt.ylabel("Y-axis", fontdict={"fontsize":20})

plt.plot(array1, label="Line 1")
plt.plot(array2, label="Line 2")
plt.plot(array3, label="Line 3")

plt.legend(shadow=True, fontsize="large")

plt.xlim(0,20) # DEFINES THE NUMBER OF TICKS THAT CAN BE PLACED IN THE X AXIS -> HERE ONLY 20 TICKS CAN BE PLACED IN THE X AXIS NOT MORE THAN THAT
plt.xticks(ticks=np.arange(1,21,1), labels=np.arange(1,21))

plt.ylim(0,400)
plt.yticks(ticks=np.arange(0,400,25), labels=np.arange(0,400,25))

print(plt.xticks())
print(plt.yticks())

plt.grid()
```

```
array4 = random.randint(10,50,10)
array5 = random.randint(20,50,10)
print(array4, array5)
```

```
plt.figure(figsize=(10,5))

plt.plot(array4, marker="o", linewidth=3)
plt.plot(array5, marker="o", linewidth=3)

plt.title("Practice graph 2",fontdict={"fontsize":20})
plt.xlabel("X-axis",fontdict={"fontsize":20})
plt.ylabel("Y-axis",fontdict={"fontsize":20})

plt.xticks(ticks=np.arange(0,10,2), labels=["Kathmandu","Lalitpur","Bhaktapur","Swoyambhu","Chhauni"], size="xx-large", rotation=45)
plt.yticks(ticks=np.arange(10,55,5), labels=["AA","BB","CC","DD","EE","FF","GG","HH","II"])

plt.grid()
```
