```
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```
x1=random.randint(10,100,5)
x1
```

```
labels = ["Milk", "Water", "Tea Leaf", "Coffee", "Sugar"]
```

```
plt.figure(figsize=(10,10))
plt.pie(x1,labels=labels, autopct='%1.1f%%', shadow=True, textprops={'fontsize':20})
plt.title("Concentration of ingridients for making milk tea", fontdict={'fontsize': 30})
```

```
x2 = random.randint(20,100,5)
x2
```

```
label2 = ["Class1", "Class2", "Class3", "Class4", "Class5"]
label2
```

```
plt.figure(figsize=(10,10))
plt.pie(x2,labels=label2, textprops={'fontsize':20}, autopct="%1.1f%%", shadow=True)
plt.title("Pass percentage in each class in a school", fontdict={'fontsize':30})
```
